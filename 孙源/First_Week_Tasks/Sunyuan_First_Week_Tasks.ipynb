```python
### Task 2 and 3 最终程序 ###
import os
path = "../" #文件夹目录
datas = []

def visitDir(filepath,subfile,hidden):   # 统计包括子文件夹(subfile判断),统计是否为隐藏hidden
    if not os.path.isdir(filepath):
        print('Error: "', filepath, '" is not a directory or does not exist.')
        return
    else:
        global filenum
        global dirnum
        global datas
        global task3
    fileNames = os.listdir(filepath)  # 获取当前路径下的文件名，返回List
    for file in fileNames:
        newDir = filepath + '/' + file # 将文件命加入到当前文件路径后面
        # print(newDir)
        # if os.path.isdir(newDir): # 如果是文件夹
        if os.path.isfile(newDir):  # 如果是文件
            if hidden:
                filenum=filenum+1
                datas.append('file '+str(filenum)+'='+newDir)
                files='file '+str(filenum)
                task3[files]=newDir
            else:
                if file[0]!='.':
                    filenum=filenum+1
                    datas.append('file '+str(filenum)+'='+newDir)
                    files='file '+str(filenum)
                    task3[files]=newDir
        else:
            if hidden:
                dirnum=dirnum+1
                datas.append('dir '+str(dirnum)+'='+newDir)
                dirs='dir '+str(dirnum)
                task3[dirs]=newDir
            else:
                if file[0]!='.':
                    dirnum=dirnum+1
                    datas.append('dir '+str(dirnum)+'='+newDir)
                    dirs='dir '+str(dirnum)
                    task3[dirs]=newDir
            if subfile:
                visitDir(newDir,subfile,hidden)                #如果不是文件，递归这个文件夹的路径

            
filenum=0
dirnum=0
datas = []
task3={}

visitDir('../',True,True)
#print(datas)
print('Task 2:')
print('   Dirs=',dirnum)
print('   Files=',filenum)
print('Task 3:')
print(task3)
```
